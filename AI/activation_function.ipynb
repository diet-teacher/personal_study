# 자연 상수 e

```python
# lim x>0
x = 0.00000001
print('(1+x)**(1/x) =', (1+x)**(1/x))
```

```python
# lim x>무한
x = 100000000
print('(1+(1/x))^x =', (1+(1/x))**x)
```

```python
print('10^(1/2.3026255437) =',10**(1/2.3026255437))
```

```python
# t가 0에 가까워지는 수 일 때
t = 0.0000001
print('10^t =', 10**t)
print('1+(2.302625*t) =', 1+(2.302625*t))
print('10^(1/2.302625) =',10**(1/2.302625))
```

```python
# t가 0에 가까워지는 수 일 때
# 두 수가 근사 값임
t = 0.0000000001
print('10^(t/2.302625) =', 10**(t/2.302625))
print('1+t =', 1+t)
```

# 활성 함수

```python
# sigmoid 함수 구현
# exp(-x)는 자연상수 e(2.7182.....)의 -x제곱을 뜻함
import numpy as np

def sigmoid(x):
    return 1/(1+2.7182817983**-x)
```

```python
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1) #y축 범위
plt.show()
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-1, 11) #y축 범위
plt.show()
```

```python
def softmax(x):
    c = np.max(x)
    exp_x = np.exp(x-c) #오버플로우 방지
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    return y
```

```python
x = np.array([0.5, 3.9, 6.1, 5.6, 1.5, 3.7])
y = softmax(x)
print(y)
```
