```python
import numpy as np
import matplotlib.pyplot as plt
```

#  함수와 그래프

```python
# 1차 함수의 그래프 3x-1

x = np.arange(-2, 2, 0.1)
a = 3
b = -1
y = a*x + b

plt.plot(x,y)
plt.plot(0, b, 'bo')
plt.axvline(x=0, color='black', linestyle='--')
plt.ylim(-10, 10)
plt.show()
```

```python
# 2차 함수의 그래프 5x^2-7

x = np.arange(-2, 2, 0.1)
a = 5
b = -7
y = a * (x**2) + b

plt.plot(x,y)
plt.plot(0, b, 'bo')
plt.axvline(x=0, color='black', linestyle='--')
plt.ylim(-8, 10)
plt.show()
```

```python
# 3차 함수의 그래프 5x^3

x = np.arange(-2, 2, 0.1)
a = 5
b = 0
y = a * (x**3) + b

plt.plot(x,y)
plt.plot(0, b, 'bo')
plt.axvline(x=0, color='black', linestyle='--')
plt.ylim(-8, 10)
plt.show()
```

# 삼각함수 그래프

```python
x = np.arange(0,10,0.01)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x,y1, label='sin')
plt.plot(x,y2, linestyle= '--', label='cos')
plt.plot(x,y3, label='tan')
plt.xticks(np.arange(0, 10, np.pi/2))
plt.legend()
plt.ylim(-2, 2)
plt.xlim(0, 10)
plt.show()
```

# 원 그래프

```python
# 중심점이 (2,4), 반지름이 3인 원 그래프

r = 3
x = np.arange(0,np.pi*2,0.01)
y = np.arange(0,np.pi*2,0.01)
a = 2
b = 4

plt.figure(figsize=(5,5))
plt.plot(a + (r*np.cos(x)), b + (r*np.sin(y)), label='loge')
plt.plot(a, b, 'bo')
plt.axvline(x=2, color='black', linestyle='--')
plt.axhline(y=4, color='black', linestyle='--')
plt.axvline(x=-1, color='red', linestyle='-')
plt.axhline(y=1, color='red', linestyle='-')
plt.axvline(x=5, color='red', linestyle='-')
plt.axhline(y=7, color='red', linestyle='-')
plt.show()
```

# sigmoid function

```python
def sigmoid(x):
    return 1/(1+np.exp(1)**-x)
    

x = np.arange(-10,10,0.1)
y1 = sigmoid(x)
y2 = sigmoid(x)*(1-sigmoid(x))

plt.plot(x,y1, label='sigmoid')
plt.plot(x,y2, label='derivative', linestyle='--')
plt.xticks(np.arange(-10, 10, 2))
plt.legend()
plt.ylim(-0.1, 1.1)
plt.xlim(-10, 10)
plt.show()
```

# hyperbolic tangent function

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.arange(-10,10,0.1)
plt.plot(x,tanh(x), label='hyperbolic tangent', color='blue')
plt.xticks(np.arange(-10, 10, 2))
plt.legend()
plt.ylim(-1.1, 1.1)
plt.xlim(-10, 10)
plt.show()
```
